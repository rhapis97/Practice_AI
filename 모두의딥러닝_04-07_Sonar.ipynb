```python
from keras.models import Sequential             # 인공신경망 층(입력층, 은닉층, 출력층) 구성
from keras.layers.core import Dense             # 전결합층(fully-conntected layer) 추가
from sklearn.preprocessing import LabelEncoder  # 카테고리형 데이터를 수치형 데이터로 변환

import pandas as pd
import numpy as np
import tensorflow as tf    # 딥러닝 프로그램을 쉽게 구현할 수 있도록 다양한 기능을 제공

# seed 값 설정
np.random.seed(3)
tf.random.set_seed(3)

# 데이터 입력
df = pd.read_csv('../dataset_deeplearning/sonar.csv', header=None)

# 데이터 개괄 보기
print(df.info())

# 데이터의 일부분 미리 보기
print(df.head())
```

```python
dataset = df.values
X = dataset[:,0:60]    # 0~59까지는 입력값
Y_obj = dataset[:,60]  # 60은 타겟값(object 형태)
```

```python
# 문자열 숫자형으로 변환
e = LabelEncoder()
e.fit(Y_obj)
Y = e.transform(Y_obj)
```

```python
# 모델 설정
model = Sequential()
model.add(Dense(24,  input_dim=60, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='sigmoid'))    # 광물과 돌을 구분하는 문제이므로 시그모이드(0, 1)를 통해 결과를 도출
```

```python
# 모델 컴파일
model.compile(loss='mean_squared_error',    # MSE는 회귀(regression) 용도 딥러닝 모델을 훈련시킬때 많이 사용되는 손실 함수
            optimizer='adam',
            metrics=['accuracy'])
```

```python
# 데이터를 배열로 변환하고 float32 형태로 타입을 바꾼다.
X = np.asarray(X).astype(np.float32)
Y = np.asarray(Y).astype(np.float32)
```

```python
# 모델 실행
model.fit(X, Y, epochs=200, batch_size=5)

# 결과 출력
print("\n Accuracy: %.4f" % (model.evaluate(X, Y)[1]))
```

정확도가 99.52%로 거의 100%에 가깝게 나온다.  
그러나 과적합일 가능성이 있다.  
따라서 과적합인지 아닌지를 확인하기 위해 학습용 데이터와 테스트용 데이터를 나누어 테스트를 해보기로 한다.

```python
from keras.models import Sequential
from keras.layers.core import Dense
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

import pandas as pd
import numpy as np
import tensorflow as tf

# seed 값 설정
seed = 0
np.random.seed(seed)
tf.random.set_seed(3)

df = pd.read_csv('../dataset_deeplearning/sonar.csv', header=None)

'''
print(df.info())
print(df.head())
'''

dataset = df.values
X = dataset[:,0:60]
Y_obj = dataset[:,60]

e = LabelEncoder()
e.fit(Y_obj)
Y = e.transform(Y_obj)
```

```python
X=np.asarray(X).astype(np.float32)
Y=np.asarray(Y).astype(np.float32)

# 학습 셋과 테스트 셋의 구분
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=seed)
# train_test_split()을 이용하여 학습 데이터와 테스트 데이터를 나눈다.
# test_size=0.3 : 테스트 데이터가 30%에 해당한다는 의미(훈련용 데이터는 70%)

model = Sequential()
model.add(Dense(24,  input_dim=60, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='mean_squared_error',
            optimizer='adam',
            metrics=['accuracy'])

model.fit(X_train, Y_train, epochs=130, batch_size=5)

# 테스트셋에 모델 적용
print("\n Test Accuracy: %.4f" % (model.evaluate(X_test, Y_test)[1]))
```

테스트셋으로 실험한 결과 정확도가 82.54%로 낮게 나왔다.  
이는 앞선 실험에서의 정확도인 99.52%보다 현저하게 낮다.

모델을 저장하고 불러와서 이용할 수 있다.

```python
from keras.models import Sequential
from keras.layers.core import Dense
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import StratifiedKFold
# StratifiedKFold: K겹 교차 검증
# 데이터가 충분하지 않을 때 데이터셋을 여러 개로 나누어 하나씩 테스트셋으로 사용하고
# 나머지를 모두 합해서 학습셋으로 사용하는 방법
# 이를 통해 가지고 있는 데이터의 100%를 테스트셋으로 사용할 수 있다.

import numpy as np
import pandas as pd
import tensorflow as tf

# seed 값 설정
seed = 0
np.random.seed(seed)
tf.random.set_seed(3)

df = pd.read_csv('../dataset_deeplearning/sonar.csv', header=None)

dataset = df.values
X = dataset[:,0:60]
Y_obj = dataset[:,60]

e = LabelEncoder()
e.fit(Y_obj)
Y = e.transform(Y_obj)

X=np.asarray(X).astype(np.float32)
Y=np.asarray(Y).astype(np.float32)
```

```python
# 10개의 파일로 쪼갬
n_fold = 10           # 자동화!(클린코딩): 언제든지 쉽게 원하는 숫자를 바꿀 수 있다.
skf = StratifiedKFold(n_splits=n_fold, shuffle=True, random_state=seed)
# n_splits: 몇 개의 파일로 쪼개 테스트할지 정한다.
# shuffle=True: 랜덤하게 데이터를 섞는다.

# 빈 accuracy 배열
accuracy = []    # n_splits가 10이므로 10개의 accuracy가 추가될 것이다.

# 모델의 설정, 컴파일, 실행
for train, test in skf.split(X, Y):    # X데이터와 Y데이터를 차례차례 집어넣어 모델을 훈련시킨다.(10번 반복)
    model = Sequential()
    model.add(Dense(24, input_dim=60, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(loss='mean_squared_error',
                  optimizer='adam',
                  metrics=['accuracy'])
    model.fit(X[train], Y[train], epochs=100, batch_size=5)        # 훈련시키기
    k_accuracy = "%.4f" % (model.evaluate(X[test], Y[test])[1])    # 테스트하기
    accuracy.append(k_accuracy)    # 테스트한 결과값을 accuracy 배열에 추가한다.

# 결과 출력
print("\n %.f fold accuracy:" % n_fold, accuracy)
```

10 fold accuracy: ['0.7143', '0.6667', '0.7619', '0.8571', '0.7619', '0.7619', '0.8571', '0.7619', '0.9000', '0.9000']  
실행 후 10번의 테스트값이 잘 출력되었다.
