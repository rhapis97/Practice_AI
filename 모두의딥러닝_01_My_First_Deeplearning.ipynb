```python
my_data = "../dataset_deeplearning/ThoraricSurgery.csv"

# -*- coding: utf-8 -*-
# 코드 내부에 한글을 사용가능 하게 해주는 부분입니다.

# 딥러닝을 구동하는 데 필요한 케라스 함수를 불러옵니다. 
# 케라스에서는 입력층, 은닉층, 출력층을 구성하기 위해 Sequential()을 사용
# Sequential()을 model로 선언한 뒤, model.add()라는 코드를 통해 층을 단계적으로 추가
from tensorflow.keras.models import Sequential
# Dense() : 전결합층(fully-conntected layer)을 추가. model.add()를 통해 추가 가능
from tensorflow.keras.layers import Dense

# 필요한 라이브러리를 불러옵니다.
import numpy as np
import tensorflow as tf

# 실행할 때마다 같은 결과를 출력하기 위해 설정하는 부분입니다.
np.random.seed(3)
tf.random.set_seed(3)

# 준비된 수술 환자 데이터를 불러들입니다.
Data_set = np.loadtxt(my_data, delimiter=",")

# 환자의 기록과 수술 결과를 X와 Y로 구분하여 저장합니다.
X = Data_set[:,0:17]    # 행 전체, 0 ~ 16열
Y = Data_set[:,17]      # 행 전체, 17열

# 딥러닝 구조를 결정합니다(모델을 설정하고 실행하는 부분입니다).
model = Sequential()
model.add(Dense(30, input_dim=17, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
# Dense()의 대표적인 인자
# 첫번째 인자: 출력 뉴런의 수
# input_dim: 입력 뉴런의 수(차원 수)
# activation: 활성화 함수
    # - linear : 기본값. 입력 뉴런과 가중치의 계산 결과를 그대로 출력한다.(선형 회귀)
    # - sigmoid : 시그모이드 함수. '이진 분류' 문제에서 출력층에 주로 사용(0과 1)
    # - softmax : 소프트맥스 함수. 셋 이상을 분류하는 '다중 클래스 분류' 문제에서 출력층에 주로 사용
    # - relu : 렐루 함수. '은닉층'에 주로 사용되는 활성화 함수.

# 딥러닝을 실행합니다.
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
# 현재 딥러닝에서 분류에 대해 가장 흔히 사용되는 손실함수는 Cross Entropy Error
# 이진 분류기를 훈련하려면, binary crossentropy 손실함수를 사용 
# 이진 분류기라는 것은 True 또는 False, 양성 또는 음성 등 2개의 클래스를 분류할 수 있는 분류기

# Adam은 Adagrad, Adadelta, RMSprop 처럼 각 파라미터마다 다른 크기의 업데이트를 적용하는 방법

model.fit(X, Y, epochs=100, batch_size=10)
# 전체 트레이닝 데이터 셋을 여러 작은 그룹을 나누었을 때 batch size는 하나의 소그룹에 속하는 데이터 수
```
